```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
np.random.seed(0)

# Generate synthetic data
data = np.random.randn(100, 3)  # 100 samples, 3 features
df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3'])

# Introduce correlation between features
df['Feature2'] = df['Feature1'] + np.random.normal(0, 0.5, 100)
df['Feature3'] = 2 * df['Feature1'] - df['Feature2'] + np.random.normal(0, 0.3, 100)

# Compute the correlation matrix
correlation_matrix = df.corr()
print("Correlation Matrix:")
print(correlation_matrix)

# Set up the matplotlib figure
plt.figure(figsize=(8, 6))

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function
import numpy as np
import pandas as pd
import plotly.graph_objs as go
import plotly.io as pio

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Step 2: Calculate z values
z = np.sin(np.sqrt(x**2 + y**2))

# Step 3: Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Step 4: Generate the 3D surface plot
surface = go.Surface(z=z, x=x, y=y)
layout = go.Layout(
    title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)
fig = go.Figure(data=[surface], layout=layout)

# Display the plot
pio.show(fig)
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Step 1: Prepare the data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Step 2: Create a new plot with a title and axis labels
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

# Step 3: Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Step 4: Display the plot in the notebook
output_notebook()
show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
import numpy as np
import pandas as pd

# Step 1: Prepare the data
np.random.seed(42)
data = {
    'x': np.random.rand(100) * 10,
    'y': np.random.rand(100) * 10,
    'sizes': np.random.randint(10, 30, 100),
    'colors': np.random.choice(['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'], 100)
}
df = pd.DataFrame(data)

# Step 2: Create a ColumnDataSource
source = ColumnDataSource(df)

# Step 3: Create a new plot with a title and axis labels
p = figure(title="Scatter Plot with Random Values", x_axis_label='x', y_axis_label='y')

# Step 4: Add a scatter renderer with varying sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', alpha=0.6, source=source)

# Step 5: Display the plot in the notebook
output_notebook()
show(p)
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral6
import pandas as pd

# Step 1: Prepare the data
data = {
    'fruits': ['Apples', 'Bananas', 'Oranges', 'Grapes', 'Pineapples'],
    'counts': [10, 20, 15, 25, 10]
}
df = pd.DataFrame(data)

# Step 2: Create a ColumnDataSource
source = ColumnDataSource(df)

# Step 3: Create a new plot with a title and axis labels
p = figure(x_range=df['fruits'], title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count', height=350, width=600)

# Step 4: Add a vbar renderer with specified width, x and top
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=Spectral6, factors=df['fruits']))

# Step 5: Customize plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Step 6: Display the plot in the notebook
output_notebook()
show(p)

```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data.
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Step 1: Generate data
data_hist = np.random.randn(1000)

# Step 2: Compute the histogram
hist, edges = np.histogram(data_hist, bins=30)

# Step 3: Prepare the data for Bokeh
source = ColumnDataSource(data=dict(
    top=hist,
    left=edges[:-1],
    right=edges[1:]
))

# Step 4: Create a new plot
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency', height=350, width=600)

# Step 5: Add a quad glyph
p.quad(top='top', bottom=0, left='left', right='right', source=source, fill_color='navy', line_color='white', alpha=0.7)

# Step 6: Customize the plot
p.y_range.start = 0
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

# Step 7: Display the plot in the notebook
output_notebook()
show(p)
```

```python
#5. Create a Bokeh heatmap using the provided dataset.
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, LinearColorMapper

# Step 1: Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Prepare data for Bokeh
x_coords = xx.flatten()
y_coords = yy.flatten()
values = data_heatmap.flatten()

# Step 2: Create a new plot
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', height=350, width=600,
           tools="hover", tooltips=[("value", "@image")])

# Step 3: Define a color mapper
color_mapper = LinearColorMapper(palette='Viridis256', low=min(values), high=max(values))

# Step 4: Add the heatmap using the image glyph
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Step 5: Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, location=(0, 0), title='Value')
p.add_layout(color_bar, 'right')

# Step 6: Display the plot in the notebook
output_notebook()
show(p)
```
